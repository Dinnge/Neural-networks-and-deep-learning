```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# 生成数据
x = np.linspace(-np.pi, np.pi, 100).reshape(-1, 1)  # 100个样本
y = np.sin(x)  # 目标函数
x_tensor = torch.FloatTensor(x)  # 转换为 PyTorch 张量
y_tensor = torch.FloatTensor(y)

# 定义模型
class TwoLayerReLU(nn.Module):
    def __init__(self):
        super(TwoLayerReLU, self).__init__()
        self.hidden = nn.Linear(1, 64)  # 第一层
        self.relu = nn.ReLU()            # ReLU 激活
        self.output = nn.Linear(64, 1)   # 输出层

    def forward(self, x):
        x = self.hidden(x)
        x = self.relu(x)
        x = self.output(x)
        return x

# 实例化模型
model = TwoLayerReLU()

# 定义损失函数和优化器
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# 训练模型
epochs = 300
for epoch in range(epochs):
    model.train()
    
    # 前向传播
    outputs = model(x_tensor)
    loss = criterion(outputs, y_tensor)

    # 反向传播和优化
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # 打印损失值
    if (epoch + 1) % 5 == 0:  # 每50轮打印一次
        print(f'Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}')

# 预测
model.eval()
with torch.no_grad():
    y_pred = model(x_tensor)

# 可视化结果
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='True Function (sin)', color='blue')
plt.plot(x, y_pred.numpy(), label='Approximated Function', color='red')
plt.title('Two-layer ReLU Network Approximating sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
