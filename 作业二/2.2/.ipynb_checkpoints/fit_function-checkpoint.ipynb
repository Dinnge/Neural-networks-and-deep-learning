```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# 
x = np.linspace(-np.pi, np.pi, 100).reshape(-1, 1)  # 100
y = np.sin(x)  # 
x_tensor = torch.FloatTensor(x)  #  PyTorch 
y_tensor = torch.FloatTensor(y)

# 
class TwoLayerReLU(nn.Module):
    def __init__(self):
        super(TwoLayerReLU, self).__init__()
        self.hidden = nn.Linear(1, 64)  # 
        self.relu = nn.ReLU()            # ReLU 
        self.output = nn.Linear(64, 1)   # 

    def forward(self, x):
        x = self.hidden(x)
        x = self.relu(x)
        x = self.output(x)
        return x

# 
model = TwoLayerReLU()

# 
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# 
epochs = 300
for epoch in range(epochs):
    model.train()
    
    # 
    outputs = model(x_tensor)
    loss = criterion(outputs, y_tensor)

    # 
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # 
    if (epoch + 1) % 5 == 0:  # 50
        print(f'Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}')

# 
model.eval()
with torch.no_grad():
    y_pred = model(x_tensor)

# 
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='True Function (sin)', color='blue')
plt.plot(x, y_pred.numpy(), label='Approximated Function', color='red')
plt.title('Two-layer ReLU Network Approximating sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
